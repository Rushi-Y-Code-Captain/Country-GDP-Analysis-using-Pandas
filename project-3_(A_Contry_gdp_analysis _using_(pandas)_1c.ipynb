```python
# Removing newly added columns 
# using df.drop() function
# syntax: obj.drop('column_name')


# drop() method can take three arguments 
# column_name: The name of the column you want to remove.
# axis=1: Specifies that you are dropping a column (use axis=0 for rows).
# inplace=True: Modifies the DataFrame directly. If you want to create a new DataFrame with the column removed,
# set inplace=False or omit this parameter.

# => column_name, axis=1, inplace = True

print(df.columns) 
# drop 
df.drop('MaCal2',axis=1,inplace= True) # remove/ dropping one column
df.drop('MyCal',axis = 1,inplace=True)
print(df.columns)

#  both the columns are removed 
# 
```
